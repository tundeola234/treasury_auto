```python
# 1. Ensure the 'Date' column is a datetime object
df2['Date'] = pd.to_datetime(df2['Date'])

# 2. Apply the filter: Year is 2025 AND Month is 11
november_2025_rows = df2[
    (df2['Date'].dt.year == 2025) &
    (df2['Date'].dt.month == 11)
]
november_2025_rows
```
